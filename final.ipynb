```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df1 = pd.read_excel("Cat.xlsx", sheet_name= "Cat4 Testing with PrepSkills (R")
df2 =pd.read_excel("Cat.xlsx",sheet_name ="Dec 31 consolidated version")
```

```python
df1 = df1.rename(columns={
    
    'Timestamp': 'timestamp',
    'Results Sent': 'results_sent',
    'With Reports ': 'with_reports',
    'Instructions sent': 'instructions_sent',
    'Email Address': 'email_address',
    'CAT4 ID Number': 'cat4_id',
    'Student FIRST name': 'first_name',
    'Student LAST name': 'last_name',
    'Student Birth MONTH': 'birth_month',
    
})

df1
```

```python
df2 = df2.rename(columns={
    'Order #': 'order_number',
    'Timestamp': 'timestamp',
    'Results Sent': 'results_sent',
    'With Reports ': 'with_reports',
    'Instructions sent': 'instructions_sent',
    'Email Address': 'email_address',
    'CAT4 ID Number': 'cat4_id',
    'Student FIRST name': 'first_name',
    'Student LAST name': 'last_name',
    'Student Birth Month': 'birth_month',
    
})
df2
```

```python
#df = pd.merge(df1, df2, on='cat4_id', how='inner')
```

```python
df = pd.concat([df1, df2],axis =1) 
```

```python
df
```

```python
df = df.loc[:, ~df.columns.duplicated()]
df
```

```python
df=df[df['timestamp'].notna()]
```

```python
df
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp']).astype('int64')
df['results_sent'] = pd.to_datetime(df['results_sent']).astype('int64')
df['Started'] = pd.to_datetime(df['Started']).astype('int64')
df['Completed'] = pd.to_datetime(df['Completed']).astype('int64')
df['instructions_sent'] = pd.to_datetime(df['instructions_sent']).astype('int64')
```

```python
df
```

```python
df = pd.get_dummies(df)
```

```python
df
```

```python
print(df['timestamp'].dtype)
print(df['instructions_sent'].dtype)
```

```python
df['timestamp'] = df['timestamp'].astype('float64')
df['results_sent'] = df['results_sent'].astype('float64')
df['Started'] = df['Started'].astype('float64')
df['Completed'] = df['Completed'].astype('float64')
df['instructions_sent']=df['instructions_sent'].astype('float64')
```

```python
x= df.drop('Percent Score', axis=1)
x = np.array(x)
```

```python
y = df['Percent Score']*100
y = np.array(y)
y
```

```python

model = LinearRegression()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r_squared = r2_score(y_test, y_pred)

print('Mean squared error: ', mse)
print('R-squared (The R2 score of our model is): ', r_squared)
```
